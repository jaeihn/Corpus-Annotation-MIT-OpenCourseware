```python
import requests

r = requests.get(url = 'https://ocw.mit.edu/courses/wgs-s10-history-of-women-in-science-and-engineering-fall-2017/pages/readings/')
#r.text
```

```python
import re

regex = r"ISBN: [0-9]*"

print("ISBN information")
print()

for match in re.finditer(regex, r.text):
    isbn = match.group(0).split()
    print(isbn)
```

```python
# Display all books information

import re
from isbntools.app import *

regex = r"ISBN: [0-9]*"

print("All books information")
print()

for match in re.finditer(regex, r.text):
    isbn = match.group(0).split()[1]
    #print(isbn)
    print(registry.bibformatters['labels'](meta(isbn, service='openl')))
    print()
```
